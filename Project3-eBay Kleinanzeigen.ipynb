```python
brand_counts=autos['brand_'].value_counts(normalize=True)
common_brand=brand_counts[brand_counts>0.05].index #这里会删选出对应的index，也就是对应的objects
```

```python
brand_mean_price={}

for c in common_brand:
    selected_dataframe=autos[autos['brand_']==c]
    mean_price=selected_dataframe['price'].mean()
    brand_mean_price[c]=int(mean_price)
brand_mean_price
```

```python
price_series=pd.Series(brand_mean_price)
print(price_series)
```

```python
brand_counts=autos['brand_'].value_counts()
brand_top20=brand_counts.head(20)
brand_top20=brand_top20.index
brand_top20
```

this is to make a new dataframe which is made by mean_price and mean_mile
dic->series->dataframe

top20 brand cars information 

```python
brand_top20_price={}
brand_top20_mile={}

for c in brand_top20:
    selected_data=autos.loc[autos.loc[:,'brand_']==c]
    selected_mean=selected_data['price'].mean()
    brand_top20_price[c]=selected_mean
    selected_mile=selected_data['odometer_km'].mean()
    brand_top20_mile[c]=selected_mile
#print(brand_top20_mile)
#print('\n')
#print(brand_top20_price)
brand_top20_price_series=pd.Series(brand_top20_price)
brand_top20_mile_series=pd.Series(brand_top20_mile)

newdataframe = pd.DataFrame(brand_top20_price_series, columns=['mean_price'])
newdataframe['mean_mile']=brand_top20_mile_series
print(newdataframe)
    
```
