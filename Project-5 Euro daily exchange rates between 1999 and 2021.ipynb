#exchange between euro and US Dollar

```python
import pandas as pd      #initialized
exchange_rates=pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.info()
exchange_rates.head()
```

#rename the columns
#format of the datetime 
#sorted with ascending 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                     'Period\\Unit:':'Time'},inplace=True)
exchange_rates['Time']=pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time',ignore_index=True,inplace=True,ascending=True)
print(exchange_rates)
```

#data cleaning 
#isolated time and us_dollar

```python
euro_to_dollar=exchange_rates[['Time','US_dollar']]
print(euro_to_dollar)
#euro_to_dollar['US_dollar'].value_counts()
euro_to_dollar=euro_to_dollar[euro_to_dollar['US_dollar']!='-']
#euro_to_dollar['US_dollar'].unique()
euro_to_dollar['US_dollar']=euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

#开始plot
#rolling mean function for long-term 

```python
import matplotlib.pyplot as plt
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
euro_to_dollar['rolling_mean']=euro_to_dollar['US_dollar'].rolling(30).mean()
plt.show()
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
euro_to_dollar
```

1. We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
    1. isolate time of 2007-2008    dt.year
    2. plot the line of 2007-2008
    3.isolate time of 2006 and 2009

```python
import matplotlib.style as style     #initialize 

style.use('fivethirtyeight')

fig,ax=plt.subplots(figsize=(8,3))   #

euro_to_dollar['Year']=euro_to_dollar['Time'].dt.year
euro_to_dollar_7_8=euro_to_dollar[euro_to_dollar['Year'].between(2007,2008)]
euro_to_dollar_6_9=euro_to_dollar[euro_to_dollar['Year'].between(2006,2009)]


ax.plot(euro_to_dollar_6_9['Time'],euro_to_dollar_6_9['rolling_mean'],color='#00FFFF',linewidth=1)

#2007-2008 with red color
ax.plot(euro_to_dollar_7_8['Time'],euro_to_dollar_7_8['rolling_mean'],color='#b00b1e')
ax.set_xticklabels([])
ax.set_yticklabels([])

print(ax.get_xticks())   #求x对应的坐标
print(ax.get_yticks())   #求y对应的坐标

## adding vertical span(rectangle) 
ax.axvspan(xmin=13950, xmax=14140, ymin=0.09,
           alpha=0.3, color='grey')

#set x 
x = 13100
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365

y= 1.19
for value in ['1.2','1.3','1.4','1.5']:
    ax.text(13010, y, value, alpha=0.5, fontsize=11)
    y+=0.1

###check the value of the peak
print(euro_to_dollar_6_9['rolling_mean'].max()) 
euro_to_dollar_6_9.sort_values('rolling_mean',ignore_index=True,inplace=True,ascending=False)

#exchange_rates.sort_values('Time',ignore_index=True,inplace=True,ascending=True)
ax.text(13950,1.6,'1.57',color='red')

ax.text(13000, 1.7,
        'Euro-USD rate peaked at 1.59 during 2007-2008\'s financial crisis',
         weight='bold')
ax.text(13000,1.65,
        'Euro-USD exchange rates between 2006 and 2010',size=14)
plt.show()

    
plt.show()
```
